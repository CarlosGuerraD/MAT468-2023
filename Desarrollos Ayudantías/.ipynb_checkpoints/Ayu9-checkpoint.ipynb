```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
unif = np.random.default_rng().uniform
normal = np.random.default_rng().normal
```

```python
k = 10000
mu_0 = 0
sigma2_0 = 1
sigma2 = 1
```

```python
mu = normal(mu_0, sigma2_0, 1)
X = normal(mu, sigma2, k)
```

```python
mu_teo = (k/sigma2 + 1/sigma2_0) * (mu_0/sigma2_0 + k*X.mean()/sigma2)
sigma2_teo = 1/(k/sigma2 + 1/sigma2_0)

X_teo = normal(mu_teo, sigma2_teo, k)
```

```python
plt.hist(X_teo, bins=40, density=True)
plt.show()
```

```python
Xt = np.zeros(k)
Xt[0] = normal(0, sigma2)



for i in range(1,k):
    Yt = normal(Xt[i-1], sigma2)
    Ut = unif(0,1)
    ratio = stats.norm.pdf(Yt, 0, sigma2) / stats.norm.pdf(Xt[-1], 0, sigma2)
    alpha = min(1, ratio)
    if Ut < alpha:
        Xt[i] = Yt
    else:
        Xt[i] = X[i-1]
```

```python
plt.hist(Xt, bins = 40)

plt.show()
```
