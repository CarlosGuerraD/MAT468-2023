## 4 Exponencial

```python
def exp_inv(x, l):
    return -(1/l)*np.log(x)
```

```python
l = 1
```

```python
array = np.random.default_rng().uniform(0,1,size=10000)
exp_array = [exp_inv(x, l) for x in array]
```

```python
plt.hist(exp_array, bins=30)
plt.show()
```

```python
stats.kstest(exp_array, stats.expon.cdf,args={l:1})
```

## 5 Weibull

```python
def exp_inv(x, l, gamma):
    return (1/l)*(-np.log(x))**(1/gamma)
```

```python
l = 1
gamma = 2
```

```python
array = np.random.default_rng().uniform(0,1,size=10000)
wei_array = [exp_inv(x, l, gamma) for x in array]
```

```python
plt.hist(wei_array, bins=30)
plt.show()
```

## 6 Bernoulli

```python
def ber_inv(x, p):
    if x <= p:
        return 0
    else:
        return 1
```

```python
p = 0.5
```
