```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import datetime
```

```python
unif = np.random.default_rng().uniform
normal = np.random.default_rng().normal
beta = np.random.default_rng().beta
poisson = np.random.default_rng().poisson
```

```python
def h_pdf(x):
    return stats.beta(2,5).pdf(x)

def g_pdf(x):
    return 2.5 * stats.uniform.pdf(x)
```

```python
def failure(n = 1):
    a = unif(0.75, 1.5)
    b = unif(2.75, 3.25)
    x = beta(a, b, size = n)
    y = 10 * x
    return y
```

```python
n = 100000
failures = 20
df = pd.DataFrame(columns=range(failures))
for i in range(1, failures + 1):
    y = np.zeros(n)
    for j in range(1, i + 1):
        y += failure(n)
    df[i] = y
```

```python
plt.style.use("default")
df.hist(figsize=(10,8),bins = 50, grid=False, density=True)
plt.suptitle("Histogram of cumulative damage for n failures")
plt.tight_layout()
plt.show()
```
